```python
import pandas as pd

path = "/content/Canada.csv"
df = pd.read_csv(path)
df
```

Total immigrants from each continent

```python
import matplotlib.pyplot as plt

df_continents = df.groupby("Continent")["Total"].sum()
plt.figure(figsize = (15, 5))
ax = df_continents.plot(kind = 'bar', color = 'orange')
plt.xlabel("Continent")
plt.ylabel("Total immigrants")
plt.title("Total immigrants from each continent")
plt.xticks(rotation = 45, ha = 'right')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.01))
plt.show()
```

Proportions of each continent in total immigrants

```python
df_continent = df.groupby('Continent')['Total'].sum().reset_index()
fig = px.pie(df_continent, values = df_continents['Total'], names = df_continent['Continent'])
fig.show()
```

Total immigrants by region.

```python
df_region = df.groupby("Region")["Total"].sum()
ax = df_region.plot(kind='bar', figsize=(16, 5))

for p in ax.patches:
    ax.annotate(
        str(int(p.get_height())),
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom',
        fontsize=10,
        rotation=0
    )

plt.ylabel("Total Immigrants")
plt.title("Total Immigrants by Region")
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

Total immigrants by developing name

```python
df_devName = df.groupby("DevName")["Total"].sum()
ax = df_devName.plot(kind='barh', figsize=(12, 6))

for p in ax.patches:
    ax.annotate(
        str(int(p.get_width())),
        (p.get_width() + 1000, p.get_y() + p.get_height() / 2),
        ha='left',
        va='center',
        fontsize=10
    )

plt.xlabel("Total Immigrants")
plt.title("Total Immigrants by developing name")
plt.tight_layout()
plt.show()
```

Immigration of top 10 countries in decades 1980s, 1990s, 2000s

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt

year_80s = list(map(str, range(1980, 1990)))
year_90s = list(map(str, range(1990, 2000)))
year_00s = list(map(str, range(2000, 2014)))
df.sort_values(['Total'], ascending = False ,inplace = True, axis = 0)
df_top10 = df.head(10)
df_80s = df_top10.loc[:, year_80s].sum(axis = 1)
df_90s = df_top10.loc[:, year_90s].sum(axis = 1)
df_00s = df_top10.loc[:, year_00s].sum(axis = 1)
new_df = pd.DataFrame({"Year_80s":df_80s, "Year_90s":df_90s, "Year_00s":df_00s})
new_df.plot(kind = 'bar', figsize = (15, 5))
plt.xlabel("Countries")
plt.ylabel("Total immigrants")
plt.title("Top 10 countries with the largest immigrants in each decade")
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

Top 10 countries has the largest immigrants

```python
df.sort_values('Total', ascending=False, inplace = True)
df_top10 = df.head(10)
ax = df_top10['Total'].plot(kind = 'bar', figsize = (15, 5))
plt.xlabel("Country")
plt.ylabel("Total")
plt.xticks(rotation = 45, ha = 'right')
for p in ax.patches:
    ax.annotate(
        str(int(p.get_height())),
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom',
        fontsize=10,
        rotation=0
    )
plt.show()
```

Top 5 countries has the largest immigrants from 1980 to 2013

```python
df.sort_values("Total", ascending=False, inplace = True)
years = list(map(str, range(1980, 2014)))
df_top5 = df.head(5)
df_top5 = df_top5[years].transpose()
df_top5.reset_index()
df_top5.plot(kind = 'area', figsize = (15, 5))
plt.xlabel("Country")
plt.ylabel("Immigrants")
```

Total immigrants from 1980 to 2013

```python
import plotly.express as px
import plotly.graph_objects as go

years = list(map(str, range(1980, 2014)))
df_total_by_year = df[years].sum(axis=0)

df_total_by_year = df_total_by_year.reset_index()
df_total_by_year.columns = ["Year", "Total"]
fig = go.Figure()
fig.add_trace(go.Scatter(x = df_total_by_year["Year"], y = df_total_by_year["Total"], mode = 'lines+markers', marker = dict(color = 'red')))
fig.update_layout(title = "Total immigrants from 1980 to 2013", xaxis_title = "Years", yaxis_title = "Total", hovermode = "x unified")
fig.show()
```

Immigrants from a specific country (example: United States)

```python
import plotly.express as px
import plotly.graph_objects as go

years = list(map(str, range(1980, 2014)))
df_new = df.loc["United States of America", years]
df_new = df_new.reset_index()
df_new.columns = ["Year", "Total"]
fig = go.Figure()
fig.add_trace(go.Scatter(x = df_new["Year"], y = df_new["Total"], mode = 'lines+markers', marker = dict(color = 'red')))
fig.update_layout(title = 'Immigrants to Canada from USA', xaxis_title = 'Year', yaxis_title = 'Total', hovermode = 'x unified')
fig.show()
```
